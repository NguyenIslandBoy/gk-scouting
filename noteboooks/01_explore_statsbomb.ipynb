```python
from statsbombpy import sb

competitions = sb.competitions()
print(competitions.shape)
competitions.head(20)
```

```python
competitions[['competition_name', 'season_name', 'competition_gender']].drop_duplicates().sort_values('competition_name')
```

```python
# La Liga 2015/16
matches = sb.matches(competition_id=11, season_id=27)
print(matches.shape)
matches[['match_id', 'home_team', 'away_team', 'match_date']].head()
```

```python
# Grab the first match_id
match_id = matches['match_id'].iloc[0]
events = sb.events(match_id=match_id)
print(events.shape)
print(events['type'].value_counts())
```

```python
gk_types = ['Shot', 'Goal Keeper', 'Pass', 'Clearance']
gk_events = events[events['type'].isin(gk_types)]
gk_events[['type', 'player', 'position', 'team']].head(20)
```

```python
gk_only = events[events['position'] == 'Goalkeeper']
print(gk_only['type'].value_counts())
gk_only[['type', 'player', 'goalkeeper_type', 'goalkeeper_outcome']].dropna(subset=['goalkeeper_type']).head(20)
```

```python
gk_events_detail = events[events['type'] == 'Goal Keeper']
print("Goalkeeper types:")
print(gk_events_detail['goalkeeper_type'].value_counts())
print("\nGoalkeeper outcomes:")
print(gk_events_detail['goalkeeper_outcome'].value_counts())
```

```python
shots = events[events['type'] == 'Shot']
print("Shot columns with data:")
print(shots[['shot_statsbomb_xg', 'shot_outcome', 'shot_technique', 'shot_body_part']].head(10))
print("\nShot outcomes:")
print(shots['shot_outcome'].value_counts())
```

```python
# PSxG is the gold standard GK metric — need to confirm it's in the data
print("PSxG column exists:", 'shot_statsbomb_xg' in events.columns)
# Also check for goalkeeper-specific xG
psxg_cols = [c for c in events.columns if 'xg' in c.lower() or 'goalkeeper' in c.lower()]
print("Relevant columns:", psxg_cols)
```
